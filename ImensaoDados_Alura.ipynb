```python
def cria_labels_feliz(labels):
                      return [label.replace("_", " ").capitalize() for label in labels]
tipos_de_tratamento = dados.tratamento.value_counts()
labels = cria_labels_feliz(dados.tratamento.unique())
explode = (0,0.2)
plt.title("Tipos de tratamento")
plt.pie(tipos_de_tratamento, labels = labels, shadow = True, explode = explode, autopct= "%1.1f%%")
plt.show()
```

```python
cod_compostos = dados["composto"].value_counts().index[0:5]
cod_compostos
```

```python
dados.query("composto in @cod_compostos") #@ é necessário quando precisamos acessar uma variável que está em outro bloco de código
```

```python
plt.figure(figsize=(8,6)) #tamanho do grafico
ax = sns.countplot(x = "composto", data = dados.query("composto in @cod_compostos"),  order = cod_compostos) #countplot plota e conta
ax.set_title("Top 5 compostos", fontsize = 18)
plt.xlabel("Composto", fontsize=16)
plt.ylabel('Contagem', fontsize=16)
plt.show()
```
